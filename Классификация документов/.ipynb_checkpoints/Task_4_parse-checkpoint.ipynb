```python
import lxml.html
import lxml.etree
import requests
import re
import numpy as np
import pandas as pd
```

```python
link = 'http://www.st-petersburg.vybory.izbirkom.ru/region/region/st-petersburg?action=show&root=1&tvd=27820001217417&vrn=27820001217413&region=78&global=&sub_region=78&prver=0&pronetvd=null&vibid=27820001217417&type=222'
```

```python
response = requests.get(link)
```

```python
tree = lxml.html.fromstring(response.text)
```

```python
tik_names = tree.xpath('//a[@style="text-decoration: none"]/text()')  # получаем названия ТИК
tik_links = tree.xpath('//a[@style="text-decoration: none"]/@href') 
```

```python
tik_names[0], tik_links[0]
```

```python
def rename(expr):
    if expr[1] == 'Территориальная избирательная комиссия':
        return 'tec' + str(expr[2])
    elif expr[1] == 'УИК':
        return 'uik' + str(expr[2])
    elif expr[0] == 'Цифровые избирательные участки':
        return 'dec'
```

```python
tik_names = list(map(lambda x: re.sub(r'(Территориальная избирательная комиссия) №(\d+)'
                                      r'|Цифровые избирательные участки',
                                      rename, x), tik_names))
tik_names[0]
```
